```python
import random
import matplotlib as plt

n = 0
N = 100000

for i in range(N):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    if (x**2 + y**2) <= 1:
        n = n + 1
print((n/N) * 4)
```

```python
sentence = "It was the best of times, it was the worst of times... in short, the period was so far like the present period, that some of its noisiest authorities insisted on its being received, for good or for evil, in the superlative degree of comparison only."
cleaned = sentence.lower().replace(',', '').replace('.', '')
properwords = cleaned.split()

word_count = {}

for word in properwords:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)
```

```python
sentence = "It was the best of times, it was the worst of times... in short, the period was so far like the present period, that some of its noisiest authorities insisted on its being received, for good or for evil, in the superlative degree of comparison only."
i = 0
result = []
for word in sentence.split(' '):
    if i % 2 == 1:
        result.append(word[::-1])
    else:
        result.append(word)
    i = i + 1

print(" ".join(result))
```
